## **2. Uji Hipotesis**

- Uji hipotesis digunakan untuk membuktikan dugaan/prediksi/hipotesa berdasarkan analisa data.
Hipotesis dibagi menjadi 2, yaitu:
  - Hipotesis null (H0)
    - It is regarding the assumption that there is no anomaly pattern or believing according to the assumption made.
    - the statements which the scientists want to prove wrong
  - Hipotesis alternatif (Ha):
    - Contrary to the null hypothesis, it shows that observation is the result of real effect.
    - statement which would be the opposite of these null hypotheses

|H0|Ha|
|---|---|
|equal (=)|not equal (≠)|
|greater than or equal to (≥) |	less than (<)|
|less than or equal to (≤)	| more than (>)|

- Terdapat beberapa uji hipotesis yang dapat dipakai, yaitu:
    - Z-test one sample
    - Z-test two sample
    - T-test one sample
    - T-test two sample
    - Chi-square

### Z-Test
Z-test digunakan pada sampel yang ukurannya lebih dari 30

#### One Sample

**Kasus**

Rata-rata kredit limit keseluruhan kostumer sebesar 8632. Apakah kredit limit kostumer perempuan lebih tinggi dibandingkan dengan rata-rata angka kredit limit seluruh kostumer? Sampel yang diambil adalah 35 data

**Hipotesis**
- H0 = Kredit limit perempuan ***lebih besar*** daripada 8632
- Ha = Kredit limit perempuan ***kurang*** dari 8632

````python
# Menghitung rata-rata keseluruhan Credit Limit
                                                                                                                      
mean_credit_limit = data["Credit_Limit"].mean()

print("Rata-rata keseluruhan Credit Limit adalah",round(mean_credit_limit))
````

````python
# Mengambil sampel kostumer perempuan sebesar 35 orang

female = data[data['Gender'] == 'F']
female
sampel_one = female['Credit_Limit'].sample(n=35)
````

````python
#Z-test

ztest, pvalue = ztest(x1=sampel_one, x2=None, value=mean_credit_limit, alternative='larger')

print ('ztest: {} and pvalue: {}'.format(ztest, pvalue))
````

````python
#Pengambilan Kesimpulan

CI = 0.95
alpha = 1-CI

if pvalue < alpha:
  print ("H0 ditolak")
  print ("pvalue {} < alpha {}".format(round(pvalue,7), round(alpha,2)))

else:
  print ("H0 diterima")
  print("pvalue {} > alpha {}".format(round(pvalue,7), round(alpha,2)))
````

Hasil uji hipotesis menggunakan Z-test didapatkan pvalue sebesar 0.999 dimana lebih besar daripada nilai alpha sehingga H0 diterima. Hal ini menunjukkan bahwa kredit limit perempuan lebih besar daripada rata-rata angka kredit keseluruhan kostumer

#### Two sample

**Kasus**
Dalam dataset the Bank Churn, Credit limit merupakan salah satu variable untuk menentukan apakah pelanggan akan churn/attrition (berhenti) atau masih tetap menggunakan produk. Lakukan hipotesis apakah rata-rata credit limit laki-laki dan perempuan sama?
- Ambil 50 samples
- Tingkat kepercayaan (conﬁdence interval) 95%
- Jenis tes mana yang anda pilih? z-test atau t-test, mengapa?
- Apakah anda menggunakan one sample atau two sample?

Hipotesis yang akan diuji adalah:
- H0 : rata-rata credit limit laki-laki sama dengan perempuan
- Ha : rata-rata credit limit laki-laki tidak sama dengan perempuan

Uji hipotesis yang digunakan adalah Z-test karena:
- Sample yang digunakan >30

Uji hipotesis ini menggunakan two sample karena membandingkan antara rata-rata credit limit laki-laki dan perempuan, yang artinya disini menggunakan dua sampel yang berbeda

````python
# Mengkategorikan sample berdasarkan jenis kelamin
laki_laki = data[data['Gender'] == 'M']
perempuan = data[data['Gender'] == 'F']

# Mengambil sampel sebanyak 50 berdasarkan masing-masing jenis kelamin
sampel_two_lk = laki_laki['Credit_Limit'].sample(n=50)
sampel_two_pr = perempuan['Credit_Limit'].sample(n=50)

#Z-test

ztest, pvalue = ztest(x1=sampel_two_lk, x2=sampel_two_pr, value=0, alternative='two-sided')

print ('ztest: {} and pvalue: {}'.format(ztest, pvalue))

#Pengambilan Kesimpulan

CI = 0.95
alpha = 1-CI

if pvalue < alpha:
  print ("H0 ditolak")
  print ("pvalue {} < alpha {}".format(round(pvalue,7), round(alpha,2)))

else:
  print ("H0 diterima")
  print("pvalue {} > alpha {}".format(round(pvalue,7), round(alpha,2)))
````

Hasil uji hipotesis menggunakan Z-test didapatkan pvalue sebesar 0.0001 dimana lebih kecil daripada nilai alpha sehingga H0 ditolak. Hal ini menunjukkan bahwa rata-rata credit limit laki-laki dan perempuan tidak sama

### T-Test

T-test digunakan pada sampel yang ukurannya kurang dari atau sama dengan 30

````python
# read dataset
data_1 = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/Transformer.csv")
data_1.head()
````

#### One Sample

**Kasus**
Apakah rata-rata oksigen yang diproduksi oleh transformer pada saat bekerja kurang dari 8357? Besar sampel yang digunakan sebanyak 20 secara acak.

**Hipotesis**
- H0 = Oksigen yang dihasilkan transformer pada saat bekerja ***lebih besar*** daripada 8357
- Ha = Oksigen yang dihasilkan transformer pada saat bekerja ***kurang*** dari 8357

````python
# Standar deviasi dan rata-rata oksigen keseluruhan data
std_oksigen = data_1['Oxigen'].std()
mean_oksigen = data_1['Oxigen'].mean()

print("Standar deviasi oksigen keseluruhan adalah:",round(std_oksigen))
print("Rata-rata oksigen keseluruhan adalah:",round(mean_oksigen))
````

````python
# Menentukan sampel
working = data_1[data_1['event'] == 'Working']
working.head()
sample_t_1 = working['Oxigen'].sample(n=20)

# ttest hypothesis
ttest_Score, pvalue = stats.ttest_1samp(sample_t_1, mean_oksigen, alternative='less') #, alternative="greater"

#Pengambilan Kesimpulan
CI = 0.95
alpha = 1-CI

if pvalue < alpha:
  print ("H0 ditolak")
  print ("pvalue {} < alpha {}".format(round(pvalue,7), round(alpha,2)))

else:
  print ("H0 diterima")
  print("pvalue {} > alpha {}".format(round(pvalue,7), round(alpha,2)))
````

Hasil uji hipotesis menggunakan T-test didapatkan pvalue sebesar 0.3182 dimana lebih besar daripada nilai alpha sehingga H0 diterima. Hal ini menunjukkan bahwa Oksigen yang dihasilkan transformer pada saat bekerja lebih besar daripada 8357

#### Two Sample

**Kasus**

Analisa apakah oksigen yang dihasilkan transformer selama bekerja dan tidak bekerja sama besar? Confidence level sebesar 95%

**Hipotesis**
- H0 : oksigen yang dihasilkan pada saat transformer bekerja ***sama dengan*** pada saat tidak bekerja
- Ha : oksigen yang dihasilkan pada saat transformer bekerja *** tidak sama dengan*** pada saat tidak bekerja

````python
# Mengkategorikan sample berdasarkan transformer bekerja
working = data_1[data_1['event']=="Working"]
mean_working_oksigen = working['Oxigen'].mean()
sampel_working_oksigen = working['Oxigen'].sample(15)

# Mengkategorikan sample berdasarkan transformer tidak bekerja
not_working = data_1[data_1['event']=="Not Working"]
mean_not_working_oksigen = not_working['Oxigen'].mean()
sampel_not_working_oksigen = not_working['Oxigen'].sample(15)

# T-Test
ttest_Score, pvalue = stats.ttest_ind(a=sampel_working_oksigen, b=sampel_not_working_oksigen, equal_var=False) #, alternative='two-sided'


#Pengambilan Kesimpulan

CI = 0.95
alpha = 1-CI

if pvalue < alpha:
  print ("H0 ditolak")
  print ("pvalue {} < alpha {}".format(round(pvalue,7), round(alpha,2)))

else:
  print ("H0 diterima")
  print("pvalue {} > alpha {}".format(round(pvalue,7), round(alpha,2)))
````

Hasil uji hipotesis menggunakan T-test didapatkan pvalue sebesar 0.1880 dimana lebih besar daripada nilai alpha sehingga H0 diterima. Hal ini menunjukkan bahwa oksigen yang dihasilkan pada saat transformer bekerja sama dengan pada saat tidak bekerja

### Chi-square

Chi square bertujuan untuk melakukan prediksi secara statistik (tes hipotesa) dengan melihat ada atau tidaknya hubungan diantara beberapa variabel

**Kasus**
Pada transformer listrik, terdapat beberapa senyawa yang dihasilkan seperti CO, CO2, Hydrogen dan Metana. Pertanyaannya adalah, apakah kegiatan transformer listrik memiliki hubungan dengan beberapa senyawa tersebut?

**Hipotesis**
- H0 = variabel tersebut berhubungan satu sama lainnya
- Ha = Variabel tersebut tidak berhubungan satu sama lainnya

````python
chi_s = data_1[['event','CO','CO2','Methane','Hydrogen']]
chi_s = chi_s.set_index('event')
chi_s.head()
````

````python
from scipy.stats import chi2_contingency
stat, pvalue, dof, expected = chi2_contingency(chi_s)

#Pengambilan Kesimpulan

CI = 0.95
alpha = 1-CI

if pvalue < alpha:
  print ("H0 ditolak")
  print ("pvalue {} < alpha {}".format(round(pvalue,7), round(alpha,2)))

else:
  print ("H0 diterima")
  print("pvalue {} > alpha {}".format(round(pvalue,7), round(alpha,2)))
````

Hasil uji hipotesis menggunakan chi-square didapatkan pvalue sebesar 00 dimana lebih kecil daripada nilai alpha sehingga H0 ditolak. Hal ini menunjukkan bahwa variabel event, CO, CO2, O2, metana, dan hidrogen tidak mempunyai hubungan satu sama lain.

```python
import pandas as pd
import numpy as np
```

## Load The Data

```python
# Load data

data = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data.csv")
print(data.shape)
data.head()
```

## **1. Sorting Data**

Pandas dapat melakukan pengurutan suatu data berdasarkan perintah dan output yang diinginkan. Misalnya:
- mengurutkan data dari yang terkecil
- mengurutkan sesuai alfabet

#### Mengambil sampel data secara random

```python
# Mengambil 7 sampel data
data.sample(n=7)
```

#### Mengurutkan data berdasarkan kolom

```python
# Secara ascending
data.sort_values(by=["region"])
```

```python
# Secara descending
data.sort_values(by=["region"], ascending=False)
```

#### Mengurutkan data berdasarkan 2 kolom

```python
data.sort_values(by=["region", "department"])
```

## **2. Filtering Data**

Pandas dapat menyeleksi beberapa data berdasarkan perintah dan output yang diinginkan. Misalnya:
- seleksi kolom tertentu
- seleksi baris dan kolom
- seleksi data tertentu pada suatu kolom


#### Mengambil beberapa kolom dari data frame

```python
data[["department", "education"]]
```

```python
data.filter(items=["department", "education"])
```

### Filtering data menggunakan `loc`

Filter data menggunakan `loc`

Fungsi `loc` untuk menyeleksi kolom dan baris berdasarkan **nama** baris dan kolom

```python
data.loc[112:117, ["education", "age"]]
```

### Filter data menggunakan `iloc`

Fungsi `iloc` untuk menyeleksi kolom dan baris berdasarkan **indeks** baris dan kolom

```python
data.iloc[-10:, 2:6]
```

## **3. Merging Data Frame**

- Merging data frame pada Pandas berfungsi untuk menggabungkan dataset berdasarkan kemiripan suatu data

- Ada 4 tipe merging data, yaitu:
  - Inner: Menampilkan baris persimpangan dari kedua dataset
  - Outer: Menampilkan semua baris dari kedua dataset
  - Left: Menampilkan semua baris dataset dari dataset sebelah kiri dan dataset yang ada disebelah kanan hanya yang cocok dengan dataset disebelah kiri
  - Right: Menampilkan semua baris dataset dari dataset sebelah kanan dan dataset yang ada disebelah kiri hanya yang cocok dengan dataset disebelah kiri


image.png

```python
# Load data2

data2 =  pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data_2.csv")
print(data2.shape)
data2.head()
```

```python
# Merging data

final = data.merge(data2, how = 'outer', on = 'employee_id')
print(final.shape)
final.head()
```

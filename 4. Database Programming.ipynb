### Kasus Ketiga

Lakukan:
- Menggabungkan table cb_station dan cb_trips berdasarkan start_station_id
- Ambil hanya kolom name, region_id, rental_methods, tripduration
- Filter tripduration di atas 1 hari (satuan tripduration adalah second)
- Lakukan pivot table guna melihat station mana aja dan tunjukan rata-rata tripduration

```python
sql_query = """
            select 
	            cs.name,
	            cs.region_id,
	            cs.rental_methods,
	            ct.tripduration 
            from batch_11.cb_stations cs
            inner join batch_11.cb_trips ct 
            on cs.station_id = ct.end_station_id
            where ct.tripduration > 86400
            order by ct.tripduration 
            """

data3 = pd.read_sql_query(sql_query, conn)
data3.head()
```

```python
# Pivot Table

data_pivot = pd.pivot_table(data3, values='tripduration', index=['name'], aggfunc='mean').round()
data_pivot
```
